```python
import csv
f = open('cig.csv', 'rb')
reader = csv.reader(f)
weight=[]
volume=[]
for row in reader:
    a =row[0]
    b = a.replace("\t", " ") 
    c = b.split()
    d = c[1:]
    e =map(float, d)   
    volume.append(e[0])
    weight.append(e[1])
print weight
print volume


f.close() 
```

```python
from pylab import *
```

```python
scatter(weight ,volume)


fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Oyster weight and volume')
ax.set_xlabel('Oyster weight (g)')
ax.set_ylabel('Oyster volume (cc)')
ax.scatter(weight, volume)
fig.savefig("oyster_with_titles.png")

```
