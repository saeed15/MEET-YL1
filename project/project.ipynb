```python
import csv 
f = open('environmental.csv', 'rb')
reader = csv.reader(f)
ozone=[]
radiation=[]

for row in reader:
    a =row[0:]
    b = a[1:] 
    d =map(float, b) 
    
    ozone.append(d[0])
    radiation.append(d[1])
   
print ozone
print radiation


f.close() 
```

```python
from pylab import *
```

```python
scatter(ozone ,radiation)


fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('environmental ozone and radiation ')
ax.set_xlabel(' environmental ozone (o)')
ax.set_ylabel('environmental  radiation(ra)')

ax.scatter(ozone, radiation,)
fig.savefig("environmental_with_titles.png")
```
